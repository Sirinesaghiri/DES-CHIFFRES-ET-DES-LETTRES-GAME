```python
import random

# We first create empty lists to store the generated letters.
letters = []

# We also prepare the list of vowels and consonants that the player can choose from.
vowels = ['a', 'e', 'i', 'o', 'u', 'y']
consonants = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 'n', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'z']

# We start a loop that will run 10 times, asking the player to choose between a vowel or a consonant.
for i in range(10):
    v_c = input("Enter 'v' for vowels or 'c' for consonants: ")
    
    # If the player chooses a vowel, we randomly select one from the list of vowels.
    if v_c == 'v':
        letter = random.choice(vowels)
        
    # If the player chooses a consonant, we randomly select one from the list of consonants.
    elif v_c == 'c':
        letter = random.choice(consonants)
        
    # We then add the selected letter (vowel or consonant) to the 'letters' list.
    letters.append(letter)

# Now we display all the letters that were generated.
print("Generated letters:", letters)

# Next, we open the 'words.txt' file that contains the dictionary of English words.
with open('words.txt', 'r') as word_file:
    dictionary = word_file.read().splitlines()  # We read the file and store the words in a list.

# This function will help us find the longest word that can be created with the generated letters.
def find_longest_word(letters, dictionary):
    longest_word = ""
    
    # We loop through each word in the dictionary and check if it can be formed with the given letters.
    for word in dictionary:
        if all(letter in letters for letter in word) and len(word) > len(longest_word):
            longest_word = word  # We update 'longest_word' if we find a longer one.
    
    return longest_word

# We ask the player to enter the word they came up with using the generated letters.
player_word = input("Enter your word: ")

# We find the longest word possible from the dictionary.
longest_word = find_longest_word(letters, dictionary)

# We display the longest word found in the dictionary.
print("Longest word found:", longest_word)

# Now we calculate the player's score for the letters round.
# We do this by comparing the length of the player's word to the length of the longest word.
L = len(longest_word)
l = len(player_word)
letter_score = max(0, 100 - (L - l) * 10)  # The score formula is applied here.
print("Your letter round score:", letter_score)

# Moving on to the numbers round, we generate a random target number between 100 and 999.
target = random.randint(100, 999)

# We also generate 6 random numbers (4 small numbers and 2 large numbers).
numbers = random.sample(range(1, 11), 4) + random.sample([25, 50, 75, 100], 2)

# We display the target number and the available numbers.
print("Target number:", target)
print("Available numbers:", numbers)

# The player is then asked to enter a mathematical expression using the available numbers to reach the target.
expression = input("Enter an expression using the available numbers to reach the target: ")

# We try to evaluate the player's expression.
result = eval(expression)  # The player's mathematical expression is evaluated.
    
    # We calculate the difference between the player's result and the target number.
difference = abs(target - result)
    
    # We then calculate the player's score for the numbers round.
number_score = max(0, 100 - (difference // 10))  # The score formula is applied here.
    
    # Finally, we display the player's result, the difference from the target, and their score.
print("Your result:", result)
print("Difference from target:", difference)
print("Your number round score:", number_score)
```
